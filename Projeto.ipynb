```python
import pandas as pd

df = pd.read_csv("best-selling-books.csv")
display(df)
```

```python
df = df.dropna()
display(df.info())
```

```python
QntLivrosPorAutores = df["Author(s)"].value_counts()
display(QntLivrosPorAutores)
PorcLivrosPorAutores = df["Author(s)"].value_counts(normalize=True)
display(PorcLivrosPorAutores)
```

```python
import plotly.express as px

graf = px.histogram(df, x="Author(s)", color="Book")
graf.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
graf.show()
```

```python
graf = px.histogram(df, x="Original language", color="Book")
graf.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
graf.show()
```

```python
for coluna in df:
    graf = px.histogram(df, x=coluna, color="Book")
    graf.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
    graf.show()
```
